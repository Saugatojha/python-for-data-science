This is different from a manual download because:
- You do not need to click, unzip, or move files yourself
- The path is returned directly into your variable, ready to use
- If you run the same code again, it loads from the local cache instantly 
  without re-downloading from the internet
- It works seamlessly on cloud environments like Google Colab or Kaggle 
  Notebooks where manual downloading is not possible

So while the files do end up stored locally, the entire process is handled 
automatically through code, making it faster, repeatable, and pipeline-friendly

locally on our system. Instead of taking 4.4 seconds to reach Kaggle's 
servers every time, kagglehub can now access the data almost instantly 
since it is already saved in the local cache folder.

```python
percentage_return_AMZN = (df.iloc[-1,1] - df.iloc[0,1]) / df.iloc[0,1] * 100
```

```python
print(percentage_return_AMZN)
```

```python
percentage_return_DPZ = (df.iloc[-1,2] - df.iloc[0,2]) / df.iloc[0,2] * 100
```

```python
print(percentage_return_DPZ)
```

```python
percentage_return_BTC = (df.iloc[-1,3] - df.iloc[0,3]) / df.iloc[0,3] * 100
```

```python
print(percentage_return_BTC)
```

```python
percentage_return_NFLX = (df.iloc[-1,4] - df.iloc[0,4]) / df.iloc[0,4] * 100
```

```python
print(percentage_return_NFLX)
```

```python
if(percentage_return_AMZN >= percentage_return_BTC >= percentage_return_DPZ >= percentage_return_NFLX):
    print(percentage_return_AMZN, "gave the highest percentage return from 2013-05-01 to 2013-05-01")

elif(percentage_return_BTC >= percentage_return_DPZ >= percentage_return_NFLX):
    print(percentage_return_BTC, "gave the highest percentage return from 2013-05-01 to 2013-05-01")

elif(percentage_return_DPZ >= percentage_return_NFLX):
    print(percentage_return_DPZ, "gave the highest percentage return from 2013-05-01 to 2013-05-01")

else:
    print(percentage_return_NFLX, "gave the highest percentage return from 2013-05-01 to 2013-05-01")
```
