kagglehub is offical python library that allows us to access/download dataset. In the next cell, I will import my dataset using kagglehub library. stored the path into an variable



```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hershyandrew/amzn-dpz-btc-ntfx-adjusted-may-2013may2019")

print("Path to dataset files:", path)
```

it is worth noting that kagglehub does save the dataset locally on your 
machine (inside a cache folder like C:\Users\admin\.cache\kagglehub\), 
it does this automatically behind the scenes with a single line of code. 

This is different from a manual download because:
- You do not need to click, unzip, or move files yourself
- The path is returned directly into your variable, ready to use
- If you run the same code again, it loads from the local cache instantly 
  without re-downloading from the internet
- It works seamlessly on cloud environments like Google Colab or Kaggle 
  Notebooks where manual downloading is not possible

So while the files do end up stored locally, the entire process is handled 
automatically through code, making it faster, repeatable, and pipeline-friendly

This is because the cache folder allows us to store frequently used data 
locally on our system. Instead of taking 4.4 seconds to reach Kaggle's 
servers every time, kagglehub can now access the data almost instantly 
since it is already saved in the local cache folder.
